```python
import json

with open('../shenzhen_cleaned.geojson', 'r', encoding='utf-8') as f:
    data = json.load(f)

print(f"总共有 {len(data['features'])} 个网格")
print(f"第一个网格的属性: {list(data['features'][0]['properties'].keys())}")
```

```python
from collections import Counter

# 统计cluster_id
cluster_ids = [feature['properties']['cluster_id'] for feature in data['features']]
cluster_counts = Counter(cluster_ids)

print(f"\n总共有 {len(cluster_counts)} 个不同的cluster_id")
print(f"各cluster_id的数量:")
for cluster_id, count in sorted(cluster_counts.items()):
    print(f"  Cluster {cluster_id}: {count} 个网格")
```

```python
from collections import Counter

# 统计 info 字段的不同取值
infos = [f["properties"].get("info") for f in data["features"]]
info_counts = Counter(infos)

print(f"总共有 {len(info_counts)} 种不同的 info 值")
print(f"\n各 info 值的数量:")
for info_val, count in info_counts.most_common():
    print(f"  {info_val}: {count} 个网格")
```

```python
import re
from collections import Counter

# 统计 info 字段的不同取值，去除括号及其中的内容
infos = [f["properties"].get("info") for f in data["features"]]
# 使用正则表达式去除 "(...)" 部分并去除多余空格
infos_cleaned = [re.sub(r'\s*\([^)]*\)', '', info).strip() for info in infos]
info_counts = Counter(infos_cleaned)

print(f"总共有 {len(info_counts)} 种不同的 info 值")
print(f"\n各 info 值的数量:")
for info_val, count in info_counts.most_common():
    print(f"  {info_val}: {count} 个网格")
```

```python
import json
import re

# 读取原始数据
with open('../grid_embeddings.geojson', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 处理每个feature
for feature in data['features']:
    props = feature['properties']
    
    # 1. 将 h3_id 重命名为 id
    if 'h3_id' in props:
        props['id'] = props.pop('h3_id')
    
    # 2. 清理 info 字段，去除括号及其内容
    if 'info' in props:
        props['info'] = re.sub(r'\s*\([^)]*\)', '', props['info']).strip()

# 保存新文件
with open('../shenzhen_cleaned.geojson', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("处理完成!")
print(f"总共处理了 {len(data['features'])} 个网格")
print(f"示例: id={data['features'][0]['properties']['id']}, info={data['features'][0]['properties']['info']}")
```
